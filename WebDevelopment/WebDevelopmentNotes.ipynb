## Adding HTML 5 
HTML 5 offers more containters to aid in logically structureing the content 
For links in the top bar, these are navigation links and there is a container for this called nav, there is no real change in the site display for this, but there will be, and the links are logically contained.   
### header 
The header represents a container for introductory content or a set of navigational links. Note: You can have several <header> elements in one HTML document. However, `<header>` cannot be placed within a `<footer>`, `<address>` or another `<header>` element. 

### article 
The `<article>` tag specifies independent, self-contained content.

An article should make sense on its own and it should be possible to distribute it independently from the rest of the site.

Potential sources for the `<article>` element:

Forum post
Blog post
News story
Note: The `<article>` element does not render as anything special in a browser. However, you can use CSS to style the `<article>` element.

### footer
This element allows for a page footer  in the example we can use web symbols the one used here is copyright a full list available from `https://www.freeformatter.com/html-entities.html`


## Semantic HTML 
One of the goals of HTML 5 was to apply more meaning to the tagging structure, in our case we exampled that by observing the use of strong and b as achieving the same aim, but strong applied a degree of meaning to the content it was working with. The other containers we used are similarly there for that reason, nav, header, footer. We could represent the same with non semantic HTML `<div>` for example, we only use the div element when we do not want to attach a certain meaning to a certain container.   
***Note*** it is important to note that the HTML5 symantics allow search engines and other applications like readers to understand the page better.
